```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Load the pre-trained Keras model
model = load_model('Best_model.keras')

# Function to preprocess the frame before prediction
def preprocess_frame(frame):
    # Resize the frame to the required size
    resized_frame = cv2.resize(frame, (48, 48))
    # Convert the frame to grayscale
    gray_frame = cv2.cvtColor(resized_frame, cv2.COLOR_BGR2GRAY)
    # Normalize the frame
    normalized_frame = gray_frame / 255.0
    # Expand dimensions to match the input shape of the model
    input_frame = np.expand_dims(normalized_frame, axis=0)
    input_frame = np.expand_dims(input_frame, axis=-1)
    return input_frame

# Open a connection to the webcam
cap = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    if not ret:
        break

    # Preprocess the frame
    input_frame = preprocess_frame(frame)

    # Make predictions
    predictions = model.predict(input_frame)

    # Process the predictions (e.g., display the predicted class)
    emotion_labels = ['anger', 'disgust', 'fear', 'happiness', 'neutral', 'sadness', 'surprise']
    predicted_emotion = emotion_labels[np.argmax(predictions)]
    cv2.putText(frame, f'Predicted Emotion: {predicted_emotion}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 2, cv2.LINE_AA)

    # Display the resulting frame
    cv2.imshow('Real-time Prediction', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
